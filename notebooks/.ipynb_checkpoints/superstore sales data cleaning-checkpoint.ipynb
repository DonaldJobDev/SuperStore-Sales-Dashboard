## Observation :
### `order_date` and `ship_date` columns contains date values but are stored as object type, because they include different formating at the same time (dd/mm/yyyy and dd-mm-yyyy)

# Converting `order_date` and `ship_date` to DateTime type üóìÔ∏è

```python
# Replace slashes with dashes
df['order_date'] = df['order_date'].str.replace('/', '-', regex=False)
df['ship_date'] = df['ship_date'].str.replace('/', '-', regex=False)

# Convert to datetime with consistent format
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True, errors='coerce')
df['ship_date'] = pd.to_datetime(df['ship_date'], dayfirst=True, errors='coerce')
```

```python
df.info()
```

```python
df.head()
```

# Convert `sales` from object type to int üî¢

```python
df['sales'].value_counts()
```

```python
# remove ',' from the values
df['sales'] = df['sales'].str.replace(',', '')
```

```python
# convert to int
df['sales'] = df['sales'].astype(int)
```

```python
df.info()
```

# Convert `year` column to DateTime üóìÔ∏è

```python
df['year'] = pd.to_datetime(df['year'], format='%Y').dt.year
```
